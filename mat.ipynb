```python
import numpy  as np 
```

```python
import matplotlib.pyplot as plt
```

```python
x= np.arange (0,10,0.1)
y= 2*x+4
fig = plt.figure(figsize = (5,5))
plt.plot(x,y,linewidth = 2,linestyle='--',color ='r',alpha =0.1,marker='+')
plt.title('line plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['line-1'],loc ='upper left')
plt.grid(True)
```

```python
x= np.arange (0,10,1)
y1= 2*x+4
y2= 3*x+10
plt.subplot(1,2,1)
plt.plot (x,y1)
plt.title('g1')
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('g2')
```

```python
x= np.arange (0,10,1)
y1= 2*x+4
y2= 3*x+10
plt.subplot(2,1,1)
plt.plot (x,y1)
plt.title('g1')
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('g2')
#side by side 
```

```python
 #scatter plot
a= np.arange(10,60,10)
b= [1,2,6,8,5]
c=[9,8,7,6,10]
plt.scatter(a,b,c='g',s=100,marker ='s',alpha=0.6)
plt.scatter(a,c,c='black',s=100,marker = 'o',alpha=1)
plt.title('yoyo')
plt.legend(['Square','Circle'])
```

```python
#box plot
#lower wisker = Q1-1.5*IQR 
#upper wisker =Q3+1.5*IQR
t= [20,4,1,30,20,10,52,50,90,30,10]
order =[10,3,1,30,20,11,30,44,2,1]
discount = [20,4,1,30,20,10,20,70]
data=  list([t,order,discount])
plt.boxplot(data)
plt.show()
```

```python
#violinplot
plt.violinplot(data,showmedians=  True)
plt.show()
```

```python
#ar
y= [10,3,1,15,17,2,30,44,22,1]
x= np.arange(0,len(y),1)
plt.stackplot(x,y,colors="black" ,alpha =0.1)
plt.plot(x,y,color="r")
plt.stackplot(x,y,alpha =1)
```

```python
#pie 
x=['dog','cats','rats','lions']
y= [40,25,10,55]
plt.pie(y,labels=x,autopct= '%0.3f%%')
```

```python
#pie 
x=['dog','cats','rats','lions']
y= [40,25,10,55]
plt.pie(y,labels=x,autopct= '%0.3f%%',explode=[0.2,0.2,0,0.0]) 
```
